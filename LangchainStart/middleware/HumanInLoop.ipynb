```python
import os 
from dotenv import load_dotenv
load_dotenv()
os.environ["Google_API_KEY"]=os.getenv("Google_API_KEY")
```

```python
from langchain.agents import create_agent
from langchain.agents.middleware import HumanInTheLoopMiddleware
from langgraph.checkpoint.memory import InMemorySaver
from langchain_core.messages import HumanMessage
from langgraph.types import Command

def read_email(email_id: str)->str:
    '''this is mock email'''
    return f'this is a content of an email {email_id}'

def send_email(recipient:str,subject:str,body:str)->str:
    '''mock function to send an email'''
    return f'the {recipient} and the  subject is {subject} and body {body}'
```

```python
agent=create_agent(
    model='google_genai:gemini-2.5-flash',
    checkpointer=InMemorySaver(),
    tools=[read_email,send_email],
    middleware=[
        HumanInTheLoopMiddleware(
            interrupt_on={
                "send_email":{
                    "allow_decisions":["approved","edit","reject"]
                },
                "read_email":False,
            }
        )
        
    ]
)
```

```python
config={"configurable":{"thread_id":"test-approved"}}

result=agent.invoke({"messages":[
    HumanMessage(
        content="send email to abc@gmail.com with hi subject and how are you body"
        )
              ]},
    config=config
)
result
```

```python
if '__interrupt__' in result:
    result=agent.invoke(
        Command(
            resume={
                'decisions':[
                    {'type':'approved'}
                ]
            }
        ),    
        config=config
    )
result
```

```python
result=agent.invoke({"messages":[
    HumanMessage(
        content="send email to abc@gmail.com with hi subject and how are you body"
        )
              ]},
    config=config
)
result
```

```python
if '__interrupt__' in result:
    result=agent.invoke(
        Command(
            resume={
                'decisions':[
                    {
                        'type':'edit',
                        'edited_action':{
                            'name':'send_email',
                            'args':{
                                'recipient':'efg@gmail.com',
                                'subject':'Hello',
                                'body':"how's going bro"
                            }
                        }
                    }
                    
                ]
            }
        ),    
        config=config
    )
result
```
